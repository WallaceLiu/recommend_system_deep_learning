* 1、取数据
* 2、构造用户与电影间的评分关系表
* 3、构造余弦相似指数并计算填充到相似矩阵中
* 4、筛选出若干相似用户
* 5、根据该用户进行电影评分推荐
* 6、输出若干推荐电影

```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
##构建用户与电影的评分矩阵
movies=pd.read_csv('./ml-latest-small/ml-latest-small/movies.csv')
ratings=pd.read_csv('./ml-latest-small/ml-latest-small/ratings.csv')
ratings_train,ratings_test=train_test_split(ratings,test_size=0.2)
data=pd.pivot_table(ratings_train[['userId','movieId','rating']],columns=['movieId'],index=['userId'],values='rating').fillna(0)
##['rating'] 会产生错误

ratings=data.values.tolist()

def cossim(x,y):
    '''余弦相似度'''
    up=0
    down1=0
    down2=0
    for (a,b) in zip(x,y):
        up+=a*b
        down1+=a**2
        down2+=b**2
    return up/(math.sqrt(down1*down2))

##构造用户相似矩阵
simmatrix=np.zeros((len(ratings),len(ratings)))
for i in range(len(ratings)-1):
    for j in range(i+1,len(ratings)):
        simmatrix[i,j]=cossim(ratings[i],ratings[j])
        simmatrix[j,i]=simmatrix[i,j]
```

```python
##为每一位用户推荐10名相似用户
user_sim_dict={}
for i in range(len(list(simmatrix))):
    user_sim_dict[i]=sorted(enumerate(list(simmatrix[i])),key=lambda x:x[1])[:10]

##对没有看过的电影打分
movie_recom_rate=np.zeros((len(ratings),len(ratings[0])))   
for i in range(len(ratings)):
    for j in range(len(ratings[i])):
        if ratings[i][j]==0:
            rate=0
            for (a,b) in user_sim_dict[i]:
                rate+=b*ratings[a][j]
            movie_recom_rate[i,j]=rate
```

```python
##每人推荐10部电影
movie_recom={}
for i in range(len(ratings)):
    movie_recom[i]=sorted(enumerate(list(movie_recom_rate[i])),key=lambda x:x[1])[:10]
```

```python
for user in movie_recom.items():
    for movie in user[1]:
        print(user_index[user[0]],movie_index[movie[0]],movies.loc[movie[0]]['title'])
```
